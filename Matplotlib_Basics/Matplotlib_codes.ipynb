```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,2,10)
y=x**2
```

## 1) Pyplot : 

```python
plt.plot(x,y,c='r', lw=5,ls='--')
```

```python
plt.scatter(x,y)
```

```python
plt.figure(figsize=(5,4))
 #On commence par cette ligne, elle represente la feuille de travail
 # figsize = le size de notre figure, les dimensions sont en inchs
plt.plot(x,y, label='quadratique')
plt.plot(x,x**3, label='cubique')
plt.title('Figure 1')
plt.xlabel('axe x')
plt.ylabel('axe y')
plt.legend()
plt.show()
plt.savefig('fig.png') #pour enregistrer la figure dans notre pc

plt.figure()
plt.plot(x,y)
```

```python
plt.subplot(2,1,1) #Grille de 2 lignes et une colonne, on va travailler sur ligne 1
plt.plot(x,y,c='r')
plt.subplot(2,1,2)
plt.plot(x,y,c='b')
```

## 2) Pyplot en OOP

```python
#on crée des objets
fig, ax=plt.subplots()
ax.plot(x,y)
plt.show()
```

```python
fig, ax=plt.subplots(2,1,sharex=False)
ax[0].plot(x,y)
ax[1].plot(x,np.sin(x))
plt.show()
```

## ***Important Code :***

```python
def graphique(data) :
    n=len(data)
    plt.figure(figsize=(12,8))
    for k, i in zip (data.keys(),range(1,n+1)) :
        plt.subplot(n,1,i)
        plt.plot(data[k])
        plt.title(k)
    plt.show()
```

## 3) Graphique de classification avec Plt.scatter :

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()

x=iris.data
y=iris.target
names=list(iris.target_names)
```

```python
plt.scatter(x[:,0], x[:,1], c=y, alpha=0.5, s=x[:,2]*100) # alpha : la transparence, s : la taille 
plt.xlabel('Longeur Sepal')
plt.ylabel('Largeur Sepal')
plt.show()
```

## 4) Graphique 3D :

```python
#%matplotlib
from mpl_toolkits.mplot3d import Axes3D
```

```python
ax=plt.axes(projection='3d') #on crée un axe dans lequel on va travailler
ax.scatter(x[:,0],x[:,1],x[:,2],c=y, s=x[:,3]*100,alpha=0.5)
```

```python
f= lambda x,y: np.sin(x)+ np.sin(y)
X=np.linspace(0,5,100)
Y=np.linspace(0,5,100)
X,Y=np.meshgrid(X,Y)
Z=f(X,Y)
ax=plt.axes(projection='3d')
ax.plot_surface(X,Y,Z,cmap='plasma')
```

## 5) Histogramme :

```python
from scipy import misc
face=misc.face(gray=True)
#plt.imshow(face,cmap='gray')
plt.hist(face.ravel(),bins=255)
plt.show()
```
