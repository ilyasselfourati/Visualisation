Exemple :

```python
titanic=sns.load_dataset('titanic')
titanic.head()
```

```python
titanic=titanic[['survived','pclass','sex','age','sibsp','fare','embarked','deck']]
titanic.dropna(axis=0,inplace=True)
titanic.head()
```

```python
sns.pairplot(titanic)
```

```python
sns.catplot(x='pclass',y='age',data=titanic,hue='sex')
```

```python
sns.boxplot(x='pclass',y='age',data=titanic,hue='sex')
```

Explication du boxplot :

<center><img src='Boxplot.png'></center>

```python
sns.distplot(titanic['fare'])
```

```python
sns.jointplot(x='age',y='fare', data=titanic, kind='hex')
```

```python
sns.heatmap(titanic.corr())
```
